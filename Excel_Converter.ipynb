```python
import pandas as pd
import os


list_name = ["NEW KAZO sale master MAR'25.xlsx",
             "NEW KAZO sale master APR'25.xlsx",
             "NEW KAZO sale master MAY'25.xlsx",
             "NEW KAZO sale master JUN'25.xlsx",
             "NEW KAZO sale master JUL'25.xlsx",
             "NEW KAZO sale master AUG'25.xlsx",
             "NEW KAZO sale master SEP'25.xlsx",
             "NEW KAZO sale master OCT'25.xlsx",
            ]
all_list = []

for fill in list_name:
    temp_read = pd.read_excel(fill,sheet_name = "MAIN DATA",engine = 'openpyxl')

    temp_read['CUSTOMER_NAME'] = temp_read['CUSTOMER_NAME'].astype(str).str.strip().str.upper() 
    temp_read = temp_read.dropna(subset = ['Date'])
    temp_read = temp_read.iloc[:,0:33]
    all_list.append(temp_read)
df_final = pd.concat(all_list,ignore_index = True)
print("Data Appened successfully")
   
```

```python
import pandas as pd


df['Date'] = pd.to_datetime(df['Date'], errors='coerce')


df = df.dropna(subset=['Date'])


df.loc[df['month'] == '#REF!', 'month'] = df['Date'].dt.strftime('%b')


month_order = ['Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct']
df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True)


df = df.sort_values('month').reset_index(drop=True)

print("Congratulation! #REF! errors are gone and months are sorted.")
print(df['month'].value_counts(sort=False))
```

```python
import pandas as pd 
import os
from openpyxl import load_workbook 

file_name = "final_net.xlsx"    
if os.path.exists(file_name):
    wb = load_workbook(file_name,read_only = True)
    sheet = wb['Sheet1']
    strt_rw = sheet.max_row
    wb.close()
    
    with pd.ExcelWriter(file_name,engine = 'openpyxl',mode = 'a',if_sheet_exists = 'overlay') as writer:
        df.to_excel(writer,index = False,sheet_name = 'Sheet1',startrow = strt_rw,header = True)
    print("Congratulation your file is append successfully")
else:
    df.to_excel(file_name,index = False, sheet_name = 'Sheet1')
    print("Congratulation your file is created successfully")
```
